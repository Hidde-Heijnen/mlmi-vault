# Linear Regression single variable

We want to minimize the following function:
$$
\sum_{i=1}^{n} (\Delta i)^2
$$


```python
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn import linear_model as lm
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../../z - Attachments/canada_per_capita_income.csv')
# printing first 3 rows for the idea
df.head(3)
```

```python
per_capita_col = 'per capita income (US$)'

plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df[per_capita_col],color='red',marker='+')
```

```python
linearRegression = lm.LinearRegression()

# training the model using the avaialble data points
linearRegression.fit(df[['year']],df[per_capita_col])
```

```python
prediction_df = pd.DataFrame({'year': [2020]})
prediction = linearRegression.predict(prediction_df)
print('Prediction for 2020: ', prediction)
```

```python
linearRegression.coef_
# gives 828.46507522
```

```python
linearRegression.intercept_
# gives -1632210.7578554575
```

```python
# linearRegression.predict([[2020]]) is the same as:
2020*linearRegression.coef_ + linearRegression.intercept_
```

```python
# plotting the linear regression line with the original data
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df[per_capita_col],color='red',marker='+')
plt.plot(df.year,linearRegression.predict(df[['year']]), color='blue')
```
